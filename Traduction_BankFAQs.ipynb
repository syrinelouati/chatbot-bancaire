```python
import argostranslate.package
import argostranslate.translate

# 1. Mettre à jour la liste des modèles disponibles
argostranslate.package.update_package_index()

# 2. Obtenir tous les modèles
available_packages = argostranslate.package.get_available_packages()

# 3. Installer le modèle EN → FR (si pas déjà installé)
try:
    en_fr_package = next(
        filter(lambda x: x.from_code == "en" and x.to_code == "fr", available_packages)
    )
    argostranslate.package.install_from_path(en_fr_package.download())
    print("Modèle EN→FR installé avec succès !")
except StopIteration:
    print("Modèle EN→FR non trouvé.")

# 4. Installer le modèle EN → AR (arabe)
try:
    en_ar_package = next(
        filter(lambda x: x.from_code == "en" and x.to_code == "ar", available_packages)
    )
    argostranslate.package.install_from_path(en_ar_package.download())
    print("Modèle EN→AR installé avec succès !")
except StopIteration:
    print("Modèle EN→AR non trouvé.")

# 5. Tester les traductions
def translate_text(text, target_lang):
    if target_lang == "fr":
        return argostranslate.translate.translate(text, "en", "fr")
    elif target_lang == "ar":
        return argostranslate.translate.translate(text, "en", "ar")
    else:
        return "Langue non supportée."

# Exemples
print("\nTraductions :")
print("Anglais → Français :", translate_text("Hello World", "fr"))
print("Anglais → Arabe :", translate_text("Hello World", "ar"))
```

# **Traduction colonne Profile**

```python
import pandas as pd
import argostranslate.translate

# Charger ton fichier CSV
df = pd.read_csv("cleaned_BankFAQs.csv")

# Définir les langues source et cible
installed_languages = argostranslate.translate.get_installed_languages()
from_lang = next(lang for lang in installed_languages if lang.code == "en")
to_lang = next(lang for lang in installed_languages if lang.code == "fr")
translation = from_lang.get_translation(to_lang)

# Traduire la colonne "text"
df["Profile_fr"] = df["Profile"].apply(lambda x: translation.translate(x) if isinstance(x, str) else x)

# Sauvegarder le résultat
df.to_csv("cleanedTranslated_BankFAQs.csv", index=False)
print("✅ Traduction terminée !")
```
